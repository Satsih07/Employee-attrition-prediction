```python
def remove_outliers(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25,0.75])
    IQR = Q3-Q1
    LE = Q1-IQR*1.5
    HE = Q3+IQR*1.5
    return LE,HE
```

```python
LE,HE = remove_outliers(df['MonthlyIncome'])
df['MonthlyIncome']=np.where(df['MonthlyIncome']>HE,HE,df['MonthlyIncome'])
df['MonthlyIncome']=np.where(df['MonthlyIncome']<LE,LE,df['MonthlyIncome'])
```

```python
LE,HE = remove_outliers(df['YearsAtCompany'])
df['YearsAtCompany']=np.where(df['YearsAtCompany']>HE,HE,df['YearsAtCompany'])
df['YearsAtCompany']=np.where(df['YearsAtCompany']<LE,LE,df['YearsAtCompany'])
```

```python
LE,HE = remove_outliers(df['YearsSinceLastPromotion'])
df['YearsSinceLastPromotion']=np.where(df['YearsSinceLastPromotion']>HE,HE,df['YearsSinceLastPromotion'])
df['YearsSinceLastPromotion']=np.where(df['YearsSinceLastPromotion']<LE,LE,df['YearsSinceLastPromotion'])
```

```python
LE,HE = remove_outliers(df['TotalWorkingYears'])
df['TotalWorkingYears']=np.where(df['TotalWorkingYears']>HE,HE,df['TotalWorkingYears'])
df['TotalWorkingYears']=np.where(df['TotalWorkingYears']<LE,LE,df['TotalWorkingYears'])
```
